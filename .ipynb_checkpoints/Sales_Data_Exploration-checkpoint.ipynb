```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
df=pd.read_csv("SampleSuperstore.csv")
df.head()
df.info()
df.describe()
df.shape
df.columns
```

```python
print("duplicate rows",df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("missing values\n:",df.isnull().sum())
df.fillna(0,inplace=True)
print("Cleaned data shape",df.shape)
```

```python
total_sales=df['Sales'].sum()
total_profit=df['Profit'].sum()
print("total sales",total_sales)
print("total profit",total_profit)
```

```python
#  GRAPHICAL REPRESENTATION OF TOTAL SALES BY REGION 
region_sales=df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
region_sales.plot(kind= 'bar',color='skyblue')
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

```python
# GRAPHICAL REPRESENTATION OF PROFITS BASED ON DIFFERENT CATEGORIES
category_profit=df.groupby('Category')['Profit'].sum()
sns.barplot(x=category_profit.index,y=category_profit.values,palette='pastel')
plt.title('Profit by Product Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```

```python
# GRAPHICAL REPRESENTATION OF TOP 5 PROFITABLE PRODUCTS
op_products=df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(5)
top_products.plot(kind= 'barh',color='skyblue')
plt.title("Top 5 profitable products")
plt.xlabel("Profit")
plt.ylabel("Product name")
plt.show()
```

```python
# GRAPHICAL REPRESENTATION OF PROFIT VS DISCOUNT
sns.scatterplot(x='Discount',y='Profit',data=df,color='red')
plt.title("profit vs discount")
plt.show()
```
